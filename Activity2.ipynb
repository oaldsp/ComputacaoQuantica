Consulte para construir, no Qiskit, um circuito qu√¢ntico que recebe o estado |0>, aplica sobre ele a porta de Hadamard, e depois faz uma medi√ß√£o do qubit resultante. Realize experimentos para plotar um histograma sobre os resultados das medi√ß√µes.


```python
from qiskit import QuantumCircuit

# Cria√ß√£o do circuito
qcircex = QuantumCircuit(1, 1)  # 1 qubit e 1 bit cl√°ssico
qcircex.h(0)                    # aplica Hadamard em |0>
qcircex.measure(0, 0)           # mede o qubit e armazena no bit cl√°ssico

# Visualizar o circuito
qcircex.draw("mpl", initial_state=True)
```

```python
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Simulador
sim = AerSimulator()

# Executa o circuito com v√°rias execu√ß√µes (shots)
job = sim.run(qcircex, shots=1000)
result = job.result()

# Obt√©m os contadores das medi√ß√µes
counts = result.get_counts()

# Plota histograma
plot_histogram(counts)
```
