Consulte para construir, no Qiskit, um circuito quântico que recebe o estado |0>, aplica sobre ele a porta de Hadamard, e depois faz uma medição do qubit resultante. Realize experimentos para plotar um histograma sobre os resultados das medições.


```python
from qiskit import QuantumCircuit

# Criação do circuito
qcircex = QuantumCircuit(1, 1)  # 1 qubit e 1 bit clássico
qcircex.h(0)                    # aplica Hadamard em |0>
qcircex.measure(0, 0)           # mede o qubit e armazena no bit clássico

# Visualizar o circuito
qcircex.draw("mpl", initial_state=True)
```

```python
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram

# Simulador
sim = AerSimulator()

# Executa o circuito com várias execuções (shots)
job = sim.run(qcircex, shots=1000)
result = job.result()

# Obtém os contadores das medições
counts = result.get_counts()

# Plota histograma
plot_histogram(counts)
```
